```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python

# Define the coefficients for the quadratic polynomial: y = ax^2 + bx + c
a = 2
b = -3
c = 1

# Generate x values
x_values = np.linspace(-10, 10, 100)  # Generate 100 x values between -10 and 10

# Calculate y values for the quadratic polynomial
y_values = a * x_values**2 + b * x_values + c

# Print the x and y values
for x, y in zip(x_values, y_values):
    print(f"x = {x}, y = {y}")
```

```python
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('x and y')
plt.show();
```

```python
x_train = x_values[:80][:, np.newaxis]
y_train = y_values[:80]
x_test = x_values[80:][:, np.newaxis]
y_test = y_values[80:]
```

```python
plt.plot(x_train, y_train)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('x and y after reshaping')
plt.show();
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
plt.plot(x_test, y_test, marker='x')
plt.plot(x_test, y_test)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('x and y')
plt.show();
```

```python
print(f'MSE: {mean_squared_error(y_true=y_test, y_pred=y_pred)}')
```
